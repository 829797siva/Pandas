 iloc[] indexer:

```python
import pandas as pd

df1 = pd.read_csv('sales1.csv')
s = df1.iloc[0]
print(s)
print()
s = df1.iloc[1]
print(s)
print()
s = df1.iloc[-1]
print(s)
print()
df2 = df1.iloc[:, 0]
print(df2)
print()
df2 = df1.iloc[:, 1]
print(df2)
print()
df2 = df1.iloc[:, -1]
print(df2)
print()
df2 = df1.iloc[0:5]
print(df2)
print()
df2 = df1.iloc[:, 0:2]
print(df2)
print()
df2 = df1.iloc[0:5, 0:2]
print(df2)
```

Setting index as a Product name and selecting  a couple of product

```python
import pandas as pd

df1 = pd.read_csv('sales4.csv')
df1.set_index("Product_Name", inplace = True)

a = ['iPhone 9', 'ThinkPad Laptop']
b = ['Product_Cost', 'Customer_Id']

df2 = df1.loc[a, b]
print(df2)
```

```python
import pandas as pd

df1 = pd.read_csv('sales4.csv')
df1.set_index("Product_Name", inplace = True)

a = ['iPhone 8', 'Google Phone']
df2 = df1.loc[a, 'Order_I Boolean / Logical indexing d' : 'Product_Cost']

print(df2)
```
