Filling NaN values with specific value

```python
import pandas as pd
import numpy as np
data = [
        ["Rajan", 26, 40000],
        ["Daniel", 16, 20000],
        ["Veeru", 45, 90000],
        ["Venkat", np.nan, 45000],
        ["Sumanth", 20, 95000],
        ["Shafi", np.nan, 97000]
]
df1 = pd.DataFrame(data, columns = ['Name', 'Age', 'Salary'])
df2 = df1.fillna(22)
print(df1)
print()
print(df2)
```

Creating dataframe and replacing nan values with specific value

```python
import pandas as pd
import numpy as np
data = [
        ["Rajan", 26, 40000],
        ["Daniel", 16, 20000],
        ["Veeru", 45, 90000],
        ["Venkat", np.nan, 45000],
        ["Sumanth", 20, 95000],
        ["Shafi", np.nan, 97000]
]
df1 = pd.DataFrame(data, columns = ['Name', 'Age', 'Salary'])
df2 = df1.replace(np.nan,0)
print(df1)
print()
print(df2)
```

Filling NaN value with mean value

```python
import pandas as pd
import numpy as np
data = [
        ["Shahid", 26, 40000],
        ["Daniel", 16, 20000],
        ["Karteek", np.nan, 90000],
        ["Venkat", np.nan, 45000],
        ["Veeru", 24, 95000],
        ["Shafi", np.nan, 97000]
]
df1 = pd.DataFrame(data, columns = ['Name', 'Age', 'Salary'])
print(df1)
m = df1['Age'].mean()
df1['Age'] = df1['Age'].fillna(m)
print()
print(df1)
```
