Converting object data type into date data type:

 We can convert from object data type into datetime data type
explicitely.

 By using,

o to_datetime(p) function

o astype(p) method in Series

2.1. to_datetime(p):

 to_datetime(p) is predefined function in pandas

 This function we should access by using pandas library.

 This function convert from object data type into date data type.

Program Loading csv file and converting Pur_Date column into Date format:

```python
import pandas as pd

df = pd.read_csv('sales7_dates.csv')

df['Pur_Date'] = pd.to_datetime(df['Pur_Date'])

print(df.head())
print()
print(df.dtypes)
```

NaT values:

 Date column may contains missing values in pandas DataFrame.

 If Date column contains missing values then while converting into Date
data type then we will get an error.

 By using errors = "coerce" keyword argument we can solve this problem.

 This argument converts Date column missing values into NaT (Not a
Time) values.

o Coerce errors i.e. convert un parse able date into NaT (Not a Time)

```python
import pandas as pd
data = {
        "Product": ["Samsung", "iPhone", "Motorola"],
        "Status": ["Success", "Success", "Failed"],
        "Cost": [10000, 50000, 15000],
        "PurDate": ['02-Sep-2019','Here date is missing','21-Sep2019']
}

df = pd.DataFrame(data)

df['PurDate'] = pd.to_datetime(df['PurDate'], errors="coerce")

print(df.head())
print()
print(df.dtypes)
```

Program Selecting Dataframe in between the dates:

 Extract year, month, day from Date column:

 Based on requirement we can get year, month, day, hour, minute from
Date column.

 Sometimes it can be useful to break up a column of dates into
components.


Program Breaking up the Date column value into multiple features :

```python
import pandas as pd

df = pd.read_csv('sales7_dates.csv', parse_dates = ['Pur_Date'])

df['year'] = df['Pur_Date'].dt.year
df['month'] = df['Pur_Date'].dt.month
df['day'] = df['Pur_Date'].dt.day

print(df.head())

```

Program Breaking up the Date column value into multiple features:

Encoding Days of the Week:

 We can get the day of the week for each date by using pandas
o Knowing the days names will helpful to understand the business
flow, like we can compare total sales on specific day for the past
three years.  


Program Encoding Days of the Week:

```python
import pandas as pd

data = {
        'Product': ['Samsung', 'iPhone', 'Motorola'],
        'Status': ['Success', 'Success', 'Failed'],
        'Cost': [10000, 50000, 15000],
        'PurDate': ['2018-01-01','2018-01-02','2018-01-03'],
}

df = pd.DataFrame(data)
df['PurDate'] = pd.to_datetime(df['PurDate'])

print(df.head())
print()
print(df["PurDate"].dt.day_name())
```
