Adding Total cost column to DataFrame

```python
import pandas as pd

df = pd.read_csv("sales8.csv")

df["Total Cost"] = df['Product cost']*df['Quantity']

print(df.head(5))
```

apply method:

```python
import pandas as pd

df = pd.read_csv('sales8.csv')

def total(df):
  t = df['Product cost'] * df['Quantity']
  return t

df['Total cost'] = df.apply(total, axis = 1)

print(df.head())
```

Program Adding total cost column to DataFrame in a specific position  

```python
import pandas as pd

df = pd.read_csv("sales8.csv")

print(df.head())

new = df['Product cost'] * df['Quantity']
df.insert(5,"Total Cost", new)

print()
print(df.head(5))
```

Dropping single columns:

```python
import pandas as pd

df1 = pd.read_csv("sales8.csv")

df2 = df1.drop(columns = 'Customer name')

print(df1.head())
print()
print(df2.head())
```

Dropping multiple columns

```python
import pandas as pd

df1 = pd.read_csv("sales8.csv")

df2 = df1.drop(['Customer name', 'Product name'], axis = 1)

print(df1.head())
print()
print(df2.head())
```

Dropping single row

```python
import pandas as pd

df1 = pd.read_csv("sales8.csv")

df2 = df1.drop(3, axis = 0)

print(df1.head())
print()
print(df2.head())
```

Dropping multiple rows


```python
import pandas as pd

df1 = pd.read_csv("sales8.csv")
df2 = df1.drop([1, 2], axis = 0)

print(df1.head())
print()
print(df2.head())
```
